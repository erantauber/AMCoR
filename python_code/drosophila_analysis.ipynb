```python
### This is the code for the drosophila control test

#important! you will need to get the homologue gene sequence for your genes of choice from NCBI
#Also the drosophila sequence text file must be named the same as the respective mammalian gene file; gene_XXX.fasta
#The files will need to be in a folder called: 'dros_gene'

#When I ran this test, I would change the files that are in the dros_gene folder when I wanted to look at higher or lower CpG genes
#It is possible to keep all the genes and run them together, before seperating them out afterwards. This is up to you.


#load packages
import os

#This fuction pulls the sequence data (sometime multiple sequences) from the inputted text files
def seqpull(file):
    with open(file, 'r') as f: 
        lines=f.read()
        lines=lines.split('\n')
        sequence=''
        sequence_dic={}
        first=0
        for line in lines:
            if first==0:
                species=line
                first=1
            elif len(line)==0:
                continue
            elif line[0]=='>':
                sequence_dic[species]=sequence
                species=line
                sequence=''
            else:
                sequence=sequence+line.upper()        
        sequence_dic[species]=sequence
        return(sequence)
    
#This counts the number of CpGs in the inputted sequence
def dyad_counter(data):
        i = 0
        CpG = 0
        dyad_locations = []
        dyad_seq=[]
        #A while loop that goes up in codons and checks between the codons for CpGs, it then counts these.
        while i< len(data):
            #counts up the index each loop
            i=i+3
            #checks if the index is above the size of the sequence after the addition. 
            if i>=len(data):
                break
        #checks if both nessecary bases are C and G, if so it adds 1 to the count. 
            if (data[i]=='G' and data[i-1]=='C'):#change this if you are not interested in dyads. 
                CpG+=1
                
        return(CpG)
    
#This automates the generation of regular gene names for later use
with open('mouse_to_human.txt','r') as f:
    lines=f.read()
    lines=lines.split('\n')
    my_dic={}
    for line in lines[1:]:
        line=line.split('\t')
        if len(line)<4 or line[2]=='' or line[3]=='':
            continue
        elif line[3] not in ens_list:
            continue
        else:
            my_dic[line[3]]=line[2]
#this pulls the general information for each gene sequence
#It is not very efficient, as it does all genes, when really we are only interested in a few.
with open('General_summary.csv', 'r') as data:
    values_dic={}
    length_dic={}
    lines = data.read()
    lines = lines.split('\n')
    file_dic={}

    for line in lines[1:]:
        line=line.split(',')
        if len(line)==1:
            continue
        else:
            values_dic[line[0]]=float(line[3])
            length_dic[line[0]]=float(line[4])
            file_dic[line[0]]=line[1]
            
            
#This creates the comparison file. I renamed the file here each time i edited the selection of genes
with open('dros_comparison_high.txt','w') as res:
    res.write('file \tspec \tval \n')
    for file in os.listdir('dros_genes/'):
        if file in file_dic.keys() and file_dic[file] in my_dic.keys():
            print(file)
            #This line pulls the sequence from the dros file
            my_seq = seqpull('dros_genes/'+file)
            #This line gets the number of CpGs in the sequence
            my_cpg=dyad_counter(my_seq)
            #These lines print out some informtion for later use, as well as write the data to a result file.
            print('Number in Dros seq: '+str(my_cpg)+'  Or at a 5th rate: '+str(my_cpg/5))
            print('Number in mammals: '+str(values_dic[file]))
            cpg1000dros = (my_cpg/len(my_seq))*1000
            print('CpGs per 1000 in Drosophila seq: '+str(cpg1000dros))
            cpg1000mal = (values_dic[file]/length_dic[file])*1000
            print('CpGs per 1000 in mammals seq: '+str(cpg1000mal))
            res.write(my_dic[file_dic[file]]+'\t'+'dros'+'\t'+str(cpg1000dros)+'\n')
            res.write(my_dic[file_dic[file]]+'\t'+'mal'+'\t'+str(cpg1000mal)+'\n')
```
